# Train_test_split

```python
from sklearn.model_selection import train_test_split
X=dataBIS.drop('stroke',axis=1)
y=dataBIS['stroke']
```

```python
dataBIS.shape
```

```python
X_Train,X_Test,y_Train,y_Test= train_test_split(X,y,test_size=0.2,random_state=10)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import SCORERS


sc=MinMaxScaler(feature_range=(0,1))
X=sc.fit_transform(X)
lg = LogisticRegression(class_weight= 'balanced')
#lg=LogisticRegression(max_iter=10000, class_weight= 'balanced')
params = {'max_iter':np.arange(50,1000,100), 'C':[.5,1,2,4]}
grid = GridSearchCV(lg, param_grid=params, scoring='recall_macro')

lg = grid.fit(X_Train, y_Train)

print(grid.best_estimator_)
#lg=lg.fit(X_Train,y_Train)
```

```python
lg = LogisticRegression(class_weight= 'balanced',max_iter=750)
lg = lg.fit(X_Train,y_Train)
```

```python
from sklearn.metrics import accuracy_score
predict1=lg.predict(X_Test)
accuracy_score(y_Test,predict1)
```

# Confusion Matrix


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import plot_confusion_matrix
cm1=confusion_matrix(y_Test,predict1)

disp = ConfusionMatrixDisplay(confusion_matrix = cm1, display_labels = lg.classes_ )
disp.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_Test, predict1))
```

```python
def create_parameters(feature, index):
    
    list_params = list()
    
    for i in range(feature):
        if(i == index):
            list_params.append(1)
        else:
            list_params.append(0)
            
    return list_params
```
